```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
from matplotlib import font_manager,rc

plt.rc("font",family = "AppleGothic")
```

```python
senior_driver = pd.read_excel("/Users/ayounglee/Desktop/multicampus/고령운전자 교통사고/가공/1823고령운전자 면허소지자수.xlsx",
                              engine = "openpyxl",header = 0)
```

```python
senior_driver = senior_driver.set_index('연도')
```

```python
senior_driver
```

```python

yx= senior_driver.plot(kind = 'bar',title="고령운전자수(면허소지자)",
                               figsize=(12, 6),
                               xlabel="연도",
                               ylabel="면허소지자수")
plt.gca().invert_xaxis()
for p in yx.patches:
    left, bottom, width, height = p.get_bbox().bounds
    # print(p.get_bbox().bounds)
    yx.annotate("%i"%(height), (left+width/2, height*1.01), ha='center', size=9)
```
