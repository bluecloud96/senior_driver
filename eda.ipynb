# 고령 운전자 면허자진반납 비율

```python
import matplotlib.pyplot as plt

years = [2022]
license_returns = [112942]
elderly_drivers = [4387358]

plt.figure(figsize=(10, 6))
bar_width = 0.3
opacity = 0.8

index = range(len(years))
bars1 = plt.bar(index, license_returns, bar_width, alpha=opacity, color='b', label='면허자진 반납 명수')
bars2 = plt.bar([i + bar_width for i in index], elderly_drivers, bar_width, alpha=opacity, label='고령운전자 면허소지자')

for bar in bars1:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), int(bar.get_height()),
             ha='center', va='bottom', fontsize=10, color='black')

for bar in bars2:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), int(bar.get_height()),
             ha='center', va='bottom', fontsize=10, color='black')

plt.xlabel('연도')
plt.ylabel('인원 수')
plt.title('2022년 면허자진 반납 명수와 고령운전자 면허소지자 수')
plt.xticks([i + bar_width / 2 for i in index], years)
plt.legend()

plt.tight_layout()
plt.show()
```

# 고령운전자수(운전면허 소지자수)

# 교통법규별 고령운전자 교통사고 건수

```python
law_violation = pd.read_excel("./data/교통법규별 고령운전자사고.xlsx",
                              engine = "openpyxl",header = 0)
```

```python
law_violation = law_violation.set_index("교통법규")
```

```python
law_violation = law_violation[["사고"]]
```

```python
import matplotlib.pyplot as plt

# 데이터 프레임을 값 기준으로 정렬합니다.
law_violation_sorted = law_violation.loc[law_violation.sum(axis=1).sort_values(ascending=False).index]

# 정렬된 데이터 프레임으로 그래프를 그립니다.
law_violation_sorted.plot(kind='bar',
                               title="고령운전자 원인별 사고수",
                               figsize=(12, 6),
                               xlabel="원인",
                               ylabel="사고수")

plt.show()
```
